## **문제를 위한 설정**

- 답이 지워지는게 싫다면 코드셀을 추가해서 사용하세요 ! 

````python
# Import Packages
import pandas as pd 
import numpy as np
import seaborn as sns

# dataset upload
df = sns.load_dataset("titanic")
df
````

## **1) index/columns 다루기**

**Q. 'survived' 컬럼을 index로 만들어 확인하고, 다시 'survived' 컬럼을 돌려놓은 뒤 인덱스를 초기화시키세요.**

````python
df.set_index('survived', inplace=True)
df
````

````python
df.reset_index(inplace=True)
df = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone', 'survived']]
df
````

**Q. DataFrame df의 컬럼명을 다음과 같이 변경하세요**
*  plcass : passenger_class &nbsp;&nbsp; | &nbsp;&nbsp; sex : gender &nbsp;&nbsp; | &nbsp;&nbsp; age : old

````python
df.rename(columns={'pclass':'passenger_class','sex':'gender','age':'old'},inplace=True)
df
````

## **2) groupby**

**Q. 'passenger_class' 값에 따른 'fare' 값의 평균, 분산, 최대, 최소 값을 구하세요.**

````python
df1 = df[['passenger_class','fare']].groupby('passenger_class').agg(['mean','var','max','min'])
df1
````

**Q. 'passenger_class' 값과 'gender' 값에 따른 'age' 의 평균을 구하세요.**

````python
df2= df.groupby(['passenger_class','gender']).old.mean()
df2
````

## **3) apply**

**Q. 'class'의 카테고리를 `apply` 함수를 이용하여, 다음과 같이 변경하세요.**  
* First : 01 &nbsp;&nbsp; | &nbsp;&nbsp; Second : 02 &nbsp;&nbsp; | &nbsp;&nbsp; Third : 03

````python
def changeCategory(x):
  if x == 'First':
    return '01'
  elif x == 'Second':
    return '02'
  elif x == 'Third':
    return '03'

df['class'] = df['class'].apply(changeCategory)
df
````

**Q. 'embark_town' 컬럼의 'Southampton'를 apply 함수를 이용하여, 'Manchester'로 변경하세요.**

````python
def changeSouthampton(x):
  if x == 'Southampton':
    return 'Manchester'
df['embark_town'] = df['embark_town'].apply(changeSouthampton)
df
````

## **4) concat/merge(합치기)**

**Q. 아래의 DataFrame df_2와 df를 하나의 프레임으로 합치세요.**

````python
# df_2 제공
data = {'col_1':100, 'col_2':200}
df_2 = pd.DataFrame(data, index=np.arange(891))

df_2
````

````python
total1 = pd.concat([df,df_2],axis=1)
total1
````

**Q. 아래의 DataFrame df_3와 df를 하나의 프레임으로 합치세요.**

````python
# df_3 제공
data = [[2, 12.0, 'S'], [1, 45.0, 'Q'], [3, 33.0, 'S']]
df_3 = pd.DataFrame(data, columns=['passenger_class', 'old', 'embarked'])

df_3
````

````python
total2= pd.merge(df,df_3)
total2
````

## **5) melt/pivot**

**Q. melt 함수를 사용하여 DataFrame df를 아래와 같은 tidy format으로 바꾸세요.**

````python
tidy = df.melt(id_vars= "survived", value_vars= ["adult_male", "alone"], var_name= "variable", value_name="Value")
tidy
````

**Q. pivot_table 함수를 사용하여, 위 문제의 tidy format을 아래와 같은 wide format으로 바꾸세요.**

````python
wide = tidy.pivot_table(index = 'survived', columns = 'variable')
wide
````
